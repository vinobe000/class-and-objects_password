```python
class password():
  def __init__(self,passwrd):
    self.passwrd=passwrd
    self.length=len(self.passwrd)
  def data(self):
    if(self.length==10) and (self.passwrd[0:3:]>=chr(97)) and (self.passwrd[0:3:]<=chr(122)) and (self.passwrd[4:6:]<=chr(57)) and (self.passwrd[4:6:]>=chr(48)) and (self.passwrd[7:8:]==chr(35)) and (self.passwrd[9::]==chr(65)):
      print(f"password is correct")
    else:
      print(f"password missing some characters")  
```

```python
class getdata(password):
  def get(self):
    print(f"{password.data(self)}")
```
